```python
import pandas as pd

def get_column_names(df):
    return df.columns.tolist()

# Assuming you have loaded your data into these DataFrames: train_data, stores_data, oil_data, and holidays_data
dataframes = {"train_data": train_data, "stores_data": stores_data, "oil_data": oil_data, "holidays_data": holidays_data}

# Get column names for each DataFrame separately
for name, df in dataframes.items():
    columns = get_column_names(df)
    print(f"Columns in {name}:")
    print(columns)
    print()  # Add an empty line for better readability
```

```python
import pandas as pd

def merge_and_show_head(df_left, df_right, on, how='left'):
    """
    Merge two DataFrames and display the head of the resulting merged DataFrame.

    Parameters:
        df_left (DataFrame): The left DataFrame.
        df_right (DataFrame): The right DataFrame.
        on (str or list of str): Column(s) to join on.
        how (str, default='left'): Type of merge to perform.

    Returns:
        merged_df (DataFrame): The merged DataFrame.
    """
    merged_df = pd.merge(df_left, df_right, on=on, how=how)
    print("Merged DataFrame:")
    print(merged_df.head())
    return merged_df

# Example usage:

# Merge train_data with oil_data on 'date'
merged_train_oil = merge_and_show_head(train_data, oil_data, on='date')

# Merge train_data with holidays_data on 'date'
merged_train_holidays = merge_and_show_head(train_data, holidays_data, on='date')

# Merge train_data with stores_data on 'store_nbr'
merged_train_stores = merge_and_show_head(train_data, stores_data, on='store_nbr')
```

```python
import pandas as pd

# Function to calculate correlation coefficient
def calculate_correlation(df, column_name):
    """
    Calculate correlation coefficient between 'sales' and a specified column.

    Parameters:
        df (DataFrame): The merged DataFrame.
        column_name (str): The name of the column to calculate correlation with 'sales'.

    Returns:
        correlation (float): The correlation coefficient.
    """
    # Calculate correlation coefficient
    correlation = None
    if len(df) > 1:
        correlation = df[['sales', column_name]].corr().iloc[0, 1]
    
    return correlation

# Assuming you have already merged your DataFrames into merged_train_oil

# Calculate correlation between sales and oil prices
correlation_oil = calculate_correlation(merged_train_oil, 'dcoilwtico')

# Print correlation coefficient
print("Correlation between sales and oil prices:", correlation_oil)
```

```python
# Calculate correlation between sales and holiday transfer status
correlation_holidays = calculate_correlation(merged_train_holidays, 'transferred')

# Print correlation coefficient
print("Correlation between sales and holiday transfer status:", correlation_holidays)
```

```python
# Calculate correlation between sales and store cluster
correlation_stores = calculate_correlation(merged_train_stores, 'cluster')

# Print correlation coefficient
print("Correlation between sales and store cluster:", correlation_stores)
```

```python
# Convert date column Â§Â§Â§ato datetime format
train_data['date'] = pd.to_datetime(train_data['date'])

# Group sales data by date and sum the sales for visualization
sales_over_time = train_data.groupby('date')['sales'].sum()

# Plot sales over time
plt.figure(figsize=(12, 6))
plt.plot(sales_over_time, color='blue')
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

```python
# Count the number of stores in each type
store_types_distribution = stores_data['type'].value_counts()

# Plot store types distribution
plt.figure(figsize=(8, 6))
store_types_distribution.plot(kind='bar', color='green')
plt.title('Store Types Distribution')
plt.xlabel('Store Type')
plt.ylabel('Number of Stores')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Convert date column to datetime format
oil_data['date'] = pd.to_datetime(oil_data['date'])

# Plot oil prices over time
plt.figure(figsize=(12, 6))
plt.plot(oil_data['date'], oil_data['dcoilwtico'], color='orange')
plt.title('Oil Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Oil Price (USD)')
plt.grid(True)
plt.show()
```

```python
# Count the number of holidays in each type
holidays_distribution = holidays_data['type'].value_counts()

# Plot holidays distribution
plt.figure(figsize=(8, 6))
holidays_distribution.plot(kind='bar', color='purple')
plt.title('Holidays Distribution')
plt.xlabel('Holiday Type')
plt.ylabel('Number of Holidays')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
